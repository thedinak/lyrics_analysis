# Writing functions to be able to repeat with more artists

```python
def get_list_of_unique_song_lyrics (artist_name):
    response = requests.get('https://www.lyrics.com/artist/'+ str(artist_name))
    page = soup(response.text, 'html.parser')
    lyrics_urls_html = page.find_all('a', attrs= {'href':re.compile('/lyric*')})
    lyrics_urls= []
    for a in lyrics_urls_html:
        lyrics_urls.append( a.get('href'))
    lyrics_urls_unique=[]
    unique_songs_first7 = []
    song_title = []
    for url in lyrics_urls:
        if str(artist_name) in str(url):
            song = url.split('/')[-1]
            if song[0:7] not in unique_songs_first7:
                unique_songs_first7.append(song[0:7])
                lyrics_urls_unique.append(url)
    return lyrics_urls_unique
    
```
